```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
db_filename = 'cereals.db'
conn = sqlite3.connect(db_filename)
c = conn.cursor()
```

```python
# Defining pie chart
c.execute("SELECT Manufacturer,count(*) FROM cereals GROUP BY Manufacturer")
counts = c.fetchall()
menuStats = pd.DataFrame.from_records(counts, columns = ['manufacturer', 'value'])
plt.pie(menuStats['value'], labels=menuStats['manufacturer'], shadow=False)
plt.axis('equal')
plt.show()
```

```python
# Defining bar chart
c.execute("SELECT Cereal, Sugars FROM cereals")
sugars = c.fetchall()
sugarFrame = pd.DataFrame.from_records(sugars, columns=['Cereals', 'Sugar Quantity'])
#print(sugarFrame)
plt.bar(range(len(sugarFrame['Sugar Quantity'])), sugarFrame['Sugar Quantity'])
plt.xticks([])
plt.show()
```

```python
#Defining histogram
c.execute("SELECT Sugars FROM cereals")
onlySugars = c.fetchall()
sugarsFrame = pd.DataFrame.from_records(onlySugars, columns=['Sugars'])
plt.hist(sugarsFrame['Sugars'], bins=9)
plt.show()
```

```python
#Defining box and whisker plot
c.execute("SELECT Sugars FROM cereals")
sugarByMan = c.fetchall()
sugarByFrame = pd.DataFrame.from_records(sugarByMan, columns=['Sugars'])
plt.boxplot(sugarByFrame['Sugars'])
plt.show()
```
